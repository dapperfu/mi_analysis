```python
import json

import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns

sns.set(
    rc={
        "axes.labelsize": 12,
        "axes.titlesize": 16,
        "figure.figsize": (10, 7.5),
        "figure.dpi": 200,
        "figure.edgecolor": "k",
        "figure.facecolor": "w",
    }
)
```

```python
df_18 = pd.read_feather("2018.ft")
df_22 = pd.read_feather("2022.ft")

df = df_18.merge(df_22, on="county")
```

```python
df
```

```python
df["gov_22_vs_18_diff"] = df["gov_22_total"] - df["gov_18_total"]
df["gov_22_vs_18_ratio"] = np.round(df["gov_22_total"] / df["gov_18_total"], 4)
```

```python
cols = [
    "county",
    "gov_18_total",
    "gov_22_total",
    "gov_22_vs_18_diff",
    "gov_22_vs_18_ratio",
]
```

```python
l = df["gov_22_vs_18_ratio"] < 1
df[l][cols]
```

```python
l = df["gov_22_vs_18_diff"] < 0
df[l][cols]
```

```python
l = df["gov_22_vs_18_diff"] > 10000
```

```python
df[l][cols].sort_values(by="gov_22_vs_18_diff", ascending=False)
```

```python
l = df["gov_22_vs_18_ratio"] > 1.12
df[l][cols].sort_values(by="gov_22_vs_18_ratio", ascending=False)
```

```python
l = df["county"] == "Ottawa"

cols = [
    "county",
    "gov_18_whitmerg",
    "gov_22_whitmer",
    "gov_18_whitmerg_pct",
    "gov_22_whitmer_pct",
    "prop3_22_yes_pct",
]

df[l][cols]
```

```python
df["gov_18_vs_22_pct_gains"] = np.round(
    df["gov_22_whitmer_pct"] - df["gov_18_whitmerg_pct"], 4
)
```

```python
cols = [
    "county",
    "gov_18_whitmerg",
    "gov_22_whitmer",
    "gov_18_whitmerg_pct",
    "gov_22_whitmer_pct",
    "gov_18_vs_22_pct_gains",
]
l = df["gov_18_vs_22_pct_gains"] > 0.04
df[l].sort_values(by="gov_18_vs_22_pct_gains")[cols]
```

```python
df["nessel_ratio"] = np.round(df["ag_22_nessel"] / df["ag_18_pct_nesseld"], 2)
df["nessel_margin"] = np.round(df["ag_22_pct_nessel"] - df["ag_18_pct_nesseld"], 2)
```

```python
l = df["nessel_ratio"] < 0.95
cols = [
    "county",
    "ag_18_nesseld",
    "ag_22_nessel",
    "ag_18_pct_nesseld",
    "ag_22_pct_nessel",
    "nessel_ratio",
    "nessel_margin",
]
```

```python
df[l][cols].sort_values(by="nessel_margin")
```

```python
l = df["nessel_margin"] > 0.08
```

```python
df[l][cols].sort_values(by="nessel_margin")
```
