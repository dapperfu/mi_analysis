```python
import json

import numpy as np
import pandas as pd
```

```python
with open("2020.general.all.json") as fp:
    races = json.load(fp)
```

```python
for idx, race in enumerate(races):
    print(f'{idx}:{race["race_name"]}')
```

```python
pres20 = races[0]
senate20 = races[1]
```

```python
data = list()
for county in senate20["counties"]:
    row = dict()
    row["county"] = county["name"]
    row["sen_20_peters"] = (
        county["results"]["petersg"] + county["results_absentee"]["petersg"]
    )
    row["sen_20_james"] = (
        county["results"]["jamesj"] + county["results_absentee"]["jamesj"]
    )
    row["sen_20_total"] = county["votes"]

    row["sen_20_peters_pct"] = np.round(row["sen_20_peters"] / county["votes"], 4)

    data.append(row)
```

```python
df_sen = pd.DataFrame(data)
```

```python
data = list()
for county in pres20["counties"]:
    row = dict()
    row["county"] = county["name"]
    row["pres_20_biden"] = (
        county["results"]["bidenj"] + county["results_absentee"]["bidenj"]
    )
    row["pres_20_trump"] = (
        county["results"]["trumpd"] + county["results_absentee"]["trumpd"]
    )
    row["pres_20_total"] = county["votes"]

    row["pres_20_biden_pct"] = np.round(row["pres_20_biden"] / county["votes"], 4)

    data.append(row)

df_pres = pd.DataFrame(data)
```

```python
df = df_pres.merge(df_sen, on="county")
```

```python
df.to_feather("2020.ft")
```
