```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score

sns.set(
    rc={
        "axes.labelsize": 12,
        "axes.titlesize": 16,
        "figure.figsize": (10, 7.5),
        "figure.dpi": 200,
        "figure.edgecolor": "k",
        "figure.facecolor": "w",
    }
)
```

```python
df = pd.read_feather("2016.ft")
df = df.merge(pd.read_feather("2018.ft"))
df = df.merge(pd.read_feather("2020.ft"))
df = df.merge(pd.read_feather("2022.ft"))
```

```python
data = list()
for idx, col in enumerate(df.columns):
    # Only compare percentages
    if "pct" not in col:
        continue
    for col2 in ["prop3_22_yes_pct"]:
        # Only compare percentages
        if col == col2:
            continue
        if "pct" not in col2:
            continue
        row = dict()
        row["election_1"] = col
        row["election_2"] = col2
        row["corr"] = df[col].corr(df[col2])
        data.append(row)
df_corr = pd.DataFrame(data)
```

```python
df3 = df_corr.sort_values(by="corr", ascending=False).reset_index(drop=True)
```
