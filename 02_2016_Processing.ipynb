```python
import json

import numpy as np
import pandas as pd
```

Primary Results Processing

```python
with open("2016.primary.json") as fp:
    races = json.load(fp)
```

```python
for idx, race in enumerate(races):
    print(f'{idx}:{race["race_id"]}')
```

```python
rep = races[0]
dem = races[1]
```

```python
# Race to data frame for the 2016 Primary.
def race_to_df(race, party):
    data = list()
    for county in race["counties"]:
        row = dict()
        row["county"] = county["name"]
        total = 0
        for key, value in county["results"].items():
            row[f'prim_16_{key.split("-")[0]}'] = value
            total += value
        row[f"prim_16_{party}_total"] = total

        for key, value in county["results"].items():
            row[f'prim_16_{key.split("-")[0]}_pct'] = np.round(value / total, 4)

        data.append(row)
    return pd.DataFrame(data)
```

```python
df_dem = race_to_df(dem, "dem")
df_rep = race_to_df(rep, "rep")
```

```python
df = df_rep.merge(df_dem, on="county")
```

General Election

```python
with open("2016.general.json") as fp:
    race2 = json.load(fp)
```

```python
df_gen = race_to_df2(race2)
```

```python
df = df.merge(df_gen, on="county")
```
